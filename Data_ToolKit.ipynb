### Question: Explanation of terms
### ***Ans***:
1. Difference between np.array, np.asarray, and np.asanyarray:
   - np.array: Always copies data unless 'copy=False' is used.
   - np.asarray: Converts input to an array, avoids copying if input is already an array.
   - np.asanyarray: Similar to np.asarray but passes subclasses (e.g., matrices).

2. Deep Copy vs Shallow Copy:
   - Deep Copy: Creates a new object and recursively copies all objects.
   - Shallow Copy: Creates a new object but references inner objects.

### Part 13: Time-series plotting with Matplotlib

```python
dates = pd.date_range('2023-01-01', periods=100)
temp = np.random.randint(20, 35, 100)
humidity = np.random.randint(30, 70, 100)
time_series_df = pd.DataFrame({'Date': dates, 'Temperature': temp, 'Humidity': humidity})
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(time_series_df['Date'], time_series_df['Temperature'], 'g-')
ax2.plot(time_series_df['Date'], time_series_df['Humidity'], 'b-')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='g')
ax2.set_ylabel('Humidity', color='b')
plt.title("Temperature and Humidity Over Time")
plt.show()
```

### 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
# Generate random arrays
x = np.random.uniform(-10, 10, 200)
y = np.random.uniform(-10, 10, 200)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = ['Q1' if (x > 0 and y > 0) else 'Q2' if (x < 0 and y > 0)
                  else 'Q3' if (x < 0 and y < 0) else 'Q4' for x, y in zip(df['x'], df['y'])]
# Scatter plot
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='tab10')
plt.axhline(0, linestyle='--', color='black')
plt.axvline(0, linestyle='--', color='black')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrant')
plt.show()
```

### 18 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8

```python
# Generate sample data
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')
temperature = np.random.uniform(20, 35, len(date_range))
humidity = np.random.uniform(40, 90, len(date_range))

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(df['Date'], df['Temperature'], 'g-', label='Temperature')
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')

# Formatting
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='green')
ax2.set_ylabel('Humidity (%)', color='blue')
plt.title('Temperature and Humidity Over Time')
plt.show()

```

### 20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'8

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import random

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(10, 100) for _ in categories]

# Create ColumnDataSource
source = ColumnDataSource(data={'categories': categories, 'values': values})

# Create plot
p = figure(x_range=categories, title='Random Categorical Bar Chart', toolbar_location=None)
p.vbar(x='categories', top='values', width=0.5, source=source, color='blue')

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show plot
show(p)
```
